```python
import pandas as pd
```

```python
games = pd.read_csv('video_games_sales.csv')  
```

```python
games
```

```python
longest_publisher = games['publisher'].dropna().str.len().idxmax()
longest_name = games.loc[longest_publisher, 'publisher']
print(f"Longest publisher name: {longest_name}, Length: {len(longest_name)}")
```

```python
games.rename(columns={'rank': 'game_id'}, inplace=True)
```

```python
games['publisher'] = games['publisher'].fillna('Unknown Publisher')
```

```python
games_with_null_publishers = games[games['publisher'].isna()]
games_with_null_publishers['publisher'] = 'Unknown Publisher'
```

```python
games_with_null_publishers
```

```python
gamesTable = games[["game_id", "name", "genre"]]
platformsTables = games[['name', 'platform', "year", "publisher", 'na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'global_sales']]
```

```python
gamesTable = gamesTable.drop_duplicates(subset='name', keep='first')
```

```python
gamesTables.value_counts()
```

```python
merged_table = platformsTables.merge(gamesTable[['game_id', 'name']], on='name', how='inner')
merged_table[merged_table['game_id'] == 253]
```

```python
gamesTables.to_csv('games_table.csv', index=False)
merged_table.to_csv('platforms_table.csv', index=False)
```
